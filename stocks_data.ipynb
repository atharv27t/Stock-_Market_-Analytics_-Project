```python
import yfinance as yf
import pandas as pd
import numpy as np
import sqlite3
import os
from datetime import datetime

```

```python
folders = ['/content/data', '/content/sql', '/content/models']

for folder in folders:
    os.makedirs(folder, exist_ok=True)
```

```python
tickers = [
    'RELIANCE.NS',
    'TCS.NS',
    'HDFCBANK.NS',
    'INFY.NS',
    'HINDUNILVR.NS',
    'ITC.NS',
    'SBIN.NS',
    'BAJFINANCE.NS',
    'ADANIENT.NS',
    'MARUTI.NS'
]

sector_map = {
    'RELIANCE.NS': 'Energy',
    'TCS.NS': 'IT',
    'HDFCBANK.NS': 'Banking',
    'INFY.NS': 'IT',
    'HINDUNILVR.NS': 'FMCG',
    'ITC.NS': 'FMCG',
    'SBIN.NS': 'Banking',
    'BAJFINANCE.NS': 'Finance',
    'ADANIENT.NS': 'IT',
    'MARUTI.NS': 'Auto'
}
```

```python
all_stocks = []

for ticker in tickers:
    print(f"Downloading {ticker}...")

    df = yf.download(
        ticker,
        start='2020-01-01',
        end='2026-06-05',
        auto_adjust=True,
        progress=False
    )


    df['ticker'] = ticker
    df['sector'] = sector_map[ticker]
    all_stocks.append(df)
```

```python
master_df = pd.concat(all_stocks)
master_df = master_df.reset_index()


master_df = master_df.rename(columns={
    'Date': 'date',
    'Open': 'open',
    'High': 'high',
    'Low': 'low',
    'Close': 'close',
    'Volume': 'volume'
})

master_df['date'] = pd.to_datetime(master_df['date'])

for col in ['open', 'high', 'low', 'close']:
    master_df[col] = master_df[col].round(2)


master_df = master_df.sort_values(['ticker', 'date']).reset_index(drop=True)

```

```python
csv_path = '/content/data/stock_prices_raw.csv'
master_df.to_csv(csv_path, index=False)

print(f"Saved: {csv_path}")
print(f"{len(master_df)} rows, {len(master_df.columns)} columns")
```
